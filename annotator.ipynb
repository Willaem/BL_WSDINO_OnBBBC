```python
import pandas
import numpy
```

# Generate Full annotated dataset

```python
df = pandas.read_csv('BBBC021_v1_image.csv')
df['batch'] = df['TableNumber'] - 1
for i, value in enumerate(numpy.sort(numpy.unique(df['Image_PathName_DAPI']))):
    df.loc[df['Image_PathName_DAPI'] == value, 'plate'] = int(i)
df['Image_Path'] = [str.split(i, sep='/')[-1] for i in df['Image_PathName_DAPI']]
df['plate'] = df['plate'].astype(int)
df.rename(columns={'Image_Metadata_Compound':'compound', 'Image_Metadata_Concentration':'compound_uM'}, inplace=True)
df.to_csv('BBBC021_full_annotated.csv')
```
